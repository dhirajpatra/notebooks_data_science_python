```python
def get_val(j, s, v=None):
    for k in j:
        if v == None and k == s:
            return j[k]
        elif v != None and k == s and v == j[k]:
            return True
        elif v != None and k == s and v != j[k]:
            return False
        elif isinstance(j[k], dict):
            return get_val(j[k], s, v)
```

```python
for l in j:
    r = get_val(l, 'mac')
    print(r)
```
